```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
url = 'Salary_Data.csv'
data = pd.read_csv(url)
data.head()
```

# Mục mới

```python
data.hist()  #ve tan suat du lieu
plt.show()
```

```python
sns.pairplot(data)  #Moi quan he giua cac cap bien
```

```python
correlation = data.corr()  #tuong quan giua cac bien(****)
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind= 'scatter', x = 'YearsExperience', y = 'Salary', figsize = (7 , 7), color = 'green')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['YearsExperience']), y= data['Salary'])
print(model.intercept_)
print(model.coef_)
print("Do chinh xac:", model.score(X = pd.DataFrame(data['YearsExperience']), y = data['Salary']))
```

```python
data.plot(kind= 'scatter', x = 'YearsExperience', y= 'Salary', figsize = (7,7), color = 'red')
y_pred = model.predict(X = pd.DataFrame(data['YearsExperience']))
plt.plot(data['YearsExperience'], y_pred, color = 'green')
```

```python
#Bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2]).T
model.fit(X, y= data['YearsExperience'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coefficient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X, y = data['Salary']))
```

```python
#Ve do thi bac 2
import numpy as np
range = np.arange(0,125000,3)
data.plot(kind= 'scatter', x = 'Salary', y= 'YearsExperience', figsize = (6,6), color = 'black')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'green')
print('Do chinh xac cua model', model.score(X, y = data['YearsExperience']))
```

```python
#bac 4
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2, data['Salary']**3, data['Salary']**4]).T
model.fit(X, y= data['YearsExperience'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coefficient cua mo hinh:', model.coef_)
print('Do chinh xac:', model.score(X, y = data['YearsExperience']))
```

```python
#Ve do thi bac 4
import numpy as np
range = np.arange(0,150000,0.5)
data.plot(kind= 'scatter', x = 'Salary', y= 'YearsExperience', figsize = (6,6), color = 'red')
predictor = pd.DataFrame([range, range**2, range**3, range**4]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'green')
print('Do chinh xac:', model.score(X, y = data['YearsExperience']))
```
